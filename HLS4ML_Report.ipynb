```python
from keras.models import load_model
model = load_model('Apr23Model.h5')
# model = load_model('KERAS_check_best_model.h5')
```

```python
config = hls4ml.utils.config_from_keras_model(model, granularity='name')

for Layer in config['LayerName'].keys():
    config['LayerName'][Layer]['Precision'] = 'ap_fixed<9,2>'   #balancing between acuracy and resource usage
    config['LayerName'][Layer]['Strategy'] = 'resource'
    config['LayerName'][Layer]['ReuseFactor'] = 30
#  how many times the resources within a layer can be reused during computation
```

```python
ls_model = hls4ml.converters.convert_from_keras_model(model,
                                                       hls_config = config,
                                                       io_type = 'io_stream',
                                                       output_dir = 'hlsProjectCreated')
```
